```python
from sklearn.preprocessing import MinMaxScaler

# Define columns for aggregated features based on their names in df_all
ac_cols = [col for col in df_all.columns if 'AC' in col and '(kW)' in col]
light_cols = [col for col in df_all.columns if 'Light' in col and '(kW)' in col]
plug_cols = [col for col in df_all.columns if 'Plug' in col and '(kW)' in col]
temp_cols = [col for col in df_all.columns if '(degC)' in col]
humidity_cols = [col for col in df_all.columns if '(RH%)' in col]
lux_cols = [col for col in df_all.columns if '(lux)' in col]

# Create new aggregated features in df_all
df_all['ac_load'] = df_all[ac_cols].sum(axis=1)
df_all['light_load'] = df_all[light_cols].sum(axis=1)
df_all['plug_load'] = df_all[plug_cols].sum(axis=1)
df_all['temp'] = df_all[temp_cols].mean(axis=1)
df_all['humidity'] = df_all[humidity_cols].mean(axis=1)
df_all['ambient_light'] = df_all[lux_cols].mean(axis=1)

# Specify the columns to be scaled (now that they are created)
scale_cols = [
    'ac_load',
    'plug_load',
    'light_load',
    'temp',
    'humidity',
    'ambient_light'
]

scaler = MinMaxScaler()

# Convert to float32 (RAM safe)
df_all[scale_cols] = df_all[scale_cols].astype('float32')

# Apply scaling
df_all[scale_cols] = scaler.fit_transform(df_all[scale_cols])

print("🟢 MinMax Scaling applied successfully to aggregated features")
df_all[scale_cols]
```

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.metrics import (
    mean_squared_error,
    mean_absolute_error,
    r2_score,
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix,
    ConfusionMatrixDisplay,
    roc_curve,
    auc
)

from tensorflow.keras.models import Model
from tensorflow.keras.layers import (
    Input,
    Dense,
    MultiHeadAttention,
    LayerNormalization,
    LSTM,
    Dropout
)

from tensorflow.keras.callbacks import (
    EarlyStopping,
    ReduceLROnPlateau
)
inputs = Input(
    shape=(1, X_train.shape[1])
)

# =========================
# Transformer Encoder
# =========================
attn_output = MultiHeadAttention(
    num_heads=2,
    key_dim=8
)(inputs, inputs)

# Project attention output to match input dimension for residual connection
attn_output_projected = Dense(X_train.shape[1])(attn_output)

x = LayerNormalization()(inputs + attn_output_projected)

# Feed Forward
ff = Dense(
    32,
    activation="relu"
)(x)

ff = Dense(
    16,
    activation="relu"
)(ff)

# Project feed-forward output to match 'x' dimension for residual connection
ff_projected = Dense(X_train.shape[1])(ff)

x = LayerNormalization()(x + ff_projected)

# =========================
# LSTM Layer
# =========================
x = LSTM(
    64,
    return_sequences=False
)(x)

x = Dropout(0.1)(x)

# =========================
# Regression Output
# =========================
output = Dense(
    1,
    activation="linear"
)(x)

hybrid_model = Model(
    inputs,
    output
)

hybrid_model.compile(
    optimizer="adam",
    loss="mse"
)

hybrid_model.summary()
```

```python
early_stop = EarlyStopping(
    monitor="val_loss",
    patience=10,
    restore_best_weights=True
)

reduce_lr = ReduceLROnPlateau(
    monitor="val_loss",
    factor=0.5,
    patience=5,
    min_lr=0.00001
)

history = hybrid_model.fit(
    X_train_tf,
    y_train,
    epochs=100,
    batch_size=16,
    validation_split=0.2,
    callbacks=[early_stop, reduce_lr],
    verbose=1
)
```

```python
y_pred = hybrid_model.predict(
    X_test_tf
).ravel()
```

```python
print("\n===== Hybrid Transformer + LSTM : Regression Metrics =====\n")

mse = mean_squared_error(
    y_test,
    y_pred
)

mae = mean_absolute_error(
    y_test,
    y_pred
)

rmse = np.sqrt(mse)

r2 = r2_score(
    y_test,
    y_pred
)

print("MSE  :", mse)
print("MAE  :", mae)
print("RMSE :", rmse)
print("R²   :", r2)
```

```python
threshold = 12.0

y_pred_class = (
    y_pred > threshold
).astype(int)
```

```python
print("\n===== Hybrid Transformer + LSTM : Classification Metrics =====\n")

accuracy = accuracy_score(
    y_test_class,
    y_pred_class
)

precision = precision_score(
    y_test_class,
    y_pred_class,
    zero_division=0
)

recall = recall_score(
    y_test_class,
    y_pred_class,
    zero_division=0
)

f1 = f1_score(
    y_test_class,
    y_pred_class,
    zero_division=0
)

cm = confusion_matrix(
    y_test_class,
    y_pred_class
)

print("Accuracy :", accuracy)
print("Precision:", precision)
print("Recall   :", recall)
print("F1-score :", f1)

print("\nConfusion Matrix:")
print(cm)
```

```python
disp = ConfusionMatrixDisplay(
    confusion_matrix=cm,
    display_labels=[
        "Low Energy",
        "High Energy"
    ]
)

disp.plot(cmap="Blues")

plt.title(
    "Hybrid Transformer + LSTM - Confusion Matrix"
)

plt.show()
```
